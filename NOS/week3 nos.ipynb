```python
import socket

def get_ip_address(domain):
    try:
        ip = socket.gethostbyname(domain)
        print(f"{domain} IP Address: {ip}")
        return ip
    except socket.gaierror:
        print(f"Could not resolve IP for {domain}")
        return None

websites = ["google.com", "facebook.com", "youtube.com"]

for site in websites:
    get_ip_address(site)
```

```python
import subprocess

def tracert(domain):
    try:
        result = subprocess.run(["tracert", domain], capture_output=True, text=True)
        print(result.stdout)
    except FileNotFoundError:
        print("Tracert command not found. Ensure it's available on your system.")
    except Exception as e:
        print(f"An error occurred: {e}")

websites = ["google.com", "facebook.com", "youtube.com"]

for site in websites:
    print(f"\nTracing route to {site}...\n")
    tracert(site)
```

```python
import subprocess
import platform

def tracert(target):
    """Runs tracert (Windows) or traceroute (Linux/macOS) for a given domain or IP."""
    try:
        command = ["tracert", target] if platform.system() == "Windows" else ["traceroute", target]

        result = subprocess.run(command, capture_output=True, text=True)
        print(result.stdout)  

    except FileNotFoundError:
        print("Trace route command not found. Ensure it's installed on your system.")
    except Exception as e:
        print(f"An error occurred: {e}")

# 3 websites
targets = [
    "twitter.com", "104.244.42.129",
    "reddit.com", "151.101.65.140",
    "linkedin.com", "13.107.42.14"
]

for target in targets:
    print(f"\nTracing route to {target}...\n")
    tracert(target)
```
